```python
import numpy as np
import skfuzzy as fuzz 
from skfuzzy import control as ctrl 
food=ctrl.Antecedent(np.arange(0,11,1),'food')
service=ctrl.Antecedent(np.arange(0,11,1),'service')
tip=ctrl.Consequent(np.arange(10,26,1),'hp')
food['poor']=fuzz.trimf(food.universe, [0,0,5])
food['average']=fuzz.trimf(food.universe, [0,5,10])
food['good']=fuzz.trimf(food.universe,[5,10,15])
service['poor']=fuzz.trimf(service.universe,[0,0,5])
service['average']=fuzz.trimf(service.universe,[0,5,10])
service['good']=fuzz.trimf(service.universe,[5,10,10])
tip['less']=fuzz.trimf(tip.universe,[10,14,18])
tip['normal']=fuzz.trimf(tip.universe,[14,18,22])
tip['much']=fuzz.trimf(tip.universe,[18,22,25])
food.view()
service.view()
tip.view()

```

```python
rule1= ctrl.Rule(food['poor'] and service['poor'],tip['less'])
rule2= ctrl.Rule(food['average'] and service['average'],tip['normal'])
rule3= ctrl.Rule(food['good'] and service['good'],tip['much'])
rule4= ctrl.Rule(food['poor'] and service['average'],tip['average'])
rule5= ctrl.Rule(food['poor'] and service['good'],tip['much'])
rule6= ctrl.Rule(food['good'] and service['average'],tip['average'])
rule7= ctrl.Rule(food['food'] and service['poor'],tip['normal'])
rule8= ctrl.Rule(food['food'] and service['average'],tip['much'])
rule9= ctrl.Rule(food['good'] and service['average'],tip['much'])
```

```python
rule1=ctrl.Rule(food['poor'] & service['poor'],tip['less'])
rule2=ctrl.Rule(food['poor'] & service['average'],tip['less'])
rule3=ctrl.Rule(food['poor'] & service['good'],tip['less'])
rule4=ctrl.Rule(food['average'] & service['poor'],tip['normal'])
rule5=ctrl.Rule(food['average'] & service['average'],tip['normal'])
rule6=ctrl.Rule(food['average'] & service['good'],tip['normal'])
rule7=ctrl.Rule(food['good'] & service['poor'],tip['much'])
rule8=ctrl.Rule(food['good'] & service['average'],tip['much'])
rule9=ctrl.Rule(food['good'] & service['good'],tip['much'])
tipping_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
tipping=ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['food']=5
tipping.input['service']=1
tipping.compute()
print(tipping.output['tip'])
tip.view(sim=tipping)
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
time = ctrl.Antecedent(np.arange(10,121,5),'time')
rice = ctrl.Antecedent(np.arange(100,1001,5),'rice')
power = ctrl.Consequent(np.arange(0,101,5),'power')
time['rat nhanh']=fuzz.trimf(time.universe,[10,10,30])
time['nhanh']=fuzz.trimf(time.universe,[10,30,60])
time['binh thuong']=fuzz.trimf(time.universe,[30,60,90])
time['cham']=fuzz.trimf(time.universe,[60,90,120])
time['rat cham']=fuzz.trimf(time.universe,[90,120,120])

rice['rat it']=fuzz.trimf(rice.universe,[100,100,300])
rice['it']=fuzz.trimf(rice.universe,[100,300,500])
rice['binh thuong']=fuzz.trimf(rice.universe,[300,500,800])
rice['nhieu']=fuzz.trimf(rice.universe,[500,800,1000])
rice['rat nhieu']=fuzz.trimf(rice.universe,[800,1000,1000])

power['rat nho']=fuzz.trimf(power.universe,[0,0,25])
power['nho']=fuzz.trimf(power.universe,[0,25,50])
power['binh thuong']=fuzz.trimf(power.universe,[25,50,75])
power['lon']=fuzz.trimf(power.universe,[50,75,100])
power['rat lon']=fuzz.trimf(power.universe,[75,100,100])

time.view()
rice.view()
power.view()

rice['binh thuong'].view()
rule1 = ctrl.Rule(time['rat nhanh']&rice['rat it'],power['rat nho'])
rule2 = ctrl.Rule(time['rat nhanh']&rice['it'],power['rat nho'])
rule3 = ctrl.Rule(time['rat nhanh']&rice['binh thuong'],power['rat nho'])
rule4 = ctrl.Rule(time['rat nhanh']&rice['nhieu'],power['nho'])
rule5 = ctrl.Rule(time['rat nhanh']&rice['rat nhieu'],power['binh thuong'])
rule6 = ctrl.Rule(time['nhanh']&rice['rat it'],power['rat nho'])
rule7 = ctrl.Rule(time['nhanh']&rice['it'],power['nho'])
rule8 = ctrl.Rule(time['nhanh']&rice['binh thuong'],power['nho'])
rule9 = ctrl.Rule(time['nhanh']&rice['nhieu'],power['binh thuong'])
rule10 = ctrl.Rule(time['nhanh']&rice['rat nhieu'],power['lon'])
rule11 = ctrl.Rule(time['binh thuong']&rice['rat it'],power['rat nho'])
rule12 = ctrl.Rule(time['binh thuong']&rice['it'],power['nho'])
rule13 = ctrl.Rule(time['binh thuong']&rice['binh thuong'],power['binh thuong'])
rule14 = ctrl.Rule(time['binh thuong']&rice['nhieu'],power['lon'])
rule15 = ctrl.Rule(time['binh thuong']&rice['rat nhieu'],power['rat lon'])
rule16 = ctrl.Rule(time['cham']&rice['rat it'],power['nho'])
rule17 = ctrl.Rule(time['cham']&rice['it'],power['binh thuong'])
rule18 = ctrl.Rule(time['cham']&rice['binh thuong'],power['lon'])
rule19 = ctrl.Rule(time['cham']&rice['nhieu'],power['lon'])
rule20 = ctrl.Rule(time['cham']&rice['rat nhieu'],power['rat lon'])
rule21 = ctrl.Rule(time['rat cham']&rice['rat it'],power['binh thuong'])
rule22 = ctrl.Rule(time['rat cham']&rice['it'],power['lon'])
rule23 = ctrl.Rule(time['rat cham']&rice['binh thuong'],power['rat lon'])
rule24 = ctrl.Rule(time['rat cham']&rice['nhieu'],power['rat lon'])
rule25 = ctrl.Rule(time['rat cham']&rice['rat nhieu'],power['rat lon'])

powering_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6, rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16, rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powering = ctrl.ControlSystemSimulation(powering_ctrl)
powering.input['time'] = 100
powering.input['rice'] = 100
powering.compute()
print(powering.output['power'])
power.view(sim=powering)
```
